```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('unnamed.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
imgbox = cv2.boxFilter(img, -1, (3,3))
plt.imshow(imgbox, cmap='gray')
```

```python
imggaus = cv2.GaussianBlur(img, (3,3) ,5)
plt.imshow(imggaus, cmap='gray')
```

```python
imgmed = cv2.medianBlur(img, 11)
plt.imshow(imgmed, cmap='gray')
```

The image becomes blurrier as the window expands. This is because as the window grows, more surrounding pixels of different values are avereaged into the center pixel's final value, leading to smaller variation in the overall pixel values.
